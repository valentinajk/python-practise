```python
def count_vowels_consonants_digits(s):
  vowels='aeiouAEIOU'
  vowel_count=sum(1 for char in s if char in vowels)
  consonant_count=sum(1 for char in s if char.isalpha() and char not in vowels)
  digit_count=sum(1 for char in s if char.isdigit())
  return vowel_count,consonant_count,digit_count
s=input("enter the string:")
vowels,consonants,digits=count_vowels_consonants_digits(s)
print(f'number of vowels : {vowels}')
print(f'number of consonants : {consonants}')
print(f'number of digits : {digits}')
```
